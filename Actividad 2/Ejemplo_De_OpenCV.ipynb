# Mi primer proyecto de vision computacional
Este primer proyecto es un ejemplo de cómo utilizar OpenCV de mandera efectiva.

```python
#Importar librerias
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
bgr = cv2.imread('city of tears.png')
print("El tamaño de la imagen es: ", bgr.shape)
#Convertir a RGB
rgb = cv2.cvtColor(bgr, cv2.COLOR_BGR2RGB)
#Mostrar la imagen
plt.imshow(rgb)
```

```python
#Mostrar la imagen en escala de grises
gray = cv2.cvtColor(bgr, cv2.COLOR_BGR2GRAY)
plt.imshow(gray, cmap='gray')
```

```python
#Mostrar las componentes de color de la imagen
plt.figure(figsize=(12,4))
plt.subplot(1,3,1)  
plt.imshow(bgr[:,:,0], cmap='gray')
plt.title('Canal Azul')
plt.subplot(1,3,2)
plt.imshow(bgr[:,:,1], cmap='gray')
plt.title('Canal Verde')
plt.subplot(1,3,3)
plt.imshow(bgr[:,:,2], cmap='gray')
plt.title('Canal Rojo')
```

```python
#Mostrar las componentes HLS de la imagen
hls = cv2.cvtColor(bgr, cv2.COLOR_BGR2HLS)
plt.figure(figsize=(12,4))
plt.subplot(1,3,1)
plt.imshow(hls[:,:,0], cmap='gray')
plt.title('Canal H')
plt.subplot(1,3,2)
plt.imshow(hls[:,:,1], cmap='gray')
plt.title('Canal L')
plt.subplot(1,3,3)
plt.imshow(hls[:,:,2], cmap='gray')
plt.title('Canal S')
plt.show()
```

```python
#Ajuste de brillo y contraste
alpha = 1.5 # Contraste
beta = 50   # Brillo

adjusted = cv2.convertScaleAbs(bgr, alpha=alpha, beta=beta)

plt.imshow(cv2.cvtColor(adjusted, cv2.COLOR_BGR2RGB))
plt.title('Imagen con brillo y contraste ajustados')
plt.show()
```

```python
#Aplicar filto Gaussiano
blurred = cv2.GaussianBlur(bgr, (15, 15), 0)
plt.imshow(cv2.cvtColor(blurred, cv2.COLOR_BGR2RGB))
plt.title('Imagen con filtro Gaussiano')
plt.show()
```
